```python
#Create data frame
import pandas as pd
df=pd.read_csv('/content/womens-world-cup.csv')
df
```

```python
df.shape
#Here 136 rows and 22 columns
```

```python
df.info()
```

```python
#I want to create a dataframe having numeric values only
df_numeric = df.select_dtypes(include = ['float64','int64'])
df_numeric
```

```python
#Input data-'minutes_played_90s' and	'goals'
#Divide the data ino input
#Slicing of input columns
x=df_numeric.iloc[:,8:10].values
x
```

```python
#Visualisation before clustering
import matplotlib.pyplot as plt
plt.scatter(df_numeric['goals'],df_numeric['minutes_played_90s'])	
```

```python
import numpy as np
np.sqrt(136)
#136 is the total number of points
#No of cluster-k
```

```python
#Find out the value of k
#Two methods are-
#1.Elbow method
from sklearn.cluster import KMeans
k=range(2,12)
#Here k range is in b/w 2 and 12
sse=[]

for i in k:
  model_demo=KMeans(n_clusters=i,random_state=0)
  model_demo.fit(x)
  sse.append(model_demo.inertia_)
plt.scatter(k,sse)
plt.plot(k,sse)
```

```python
#2.Silhouette score method to find out k value
from sklearn.metrics import silhouette_score
k=range(2,12)
for i in k:
  model_demo=KMeans(n_clusters=i,random_state=0)
  model_demo.fit(x)
  y_pred=model_demo.predict(x)
  print(f"{i} Clusters ,Score={silhouette_score(x,y_pred)}")
  plt.bar(i,silhouette_score(x,y_pred))
```

```python
#Apply clusterer
k=2
from sklearn.cluster import KMeans
model=KMeans(n_clusters=k,random_state=0)
model.fit(x)
```

```python
#Output prediction
y=model.predict(x)
y
```

```python
y.size
```

```python
x[y==0,1]
#the value of input when cluster 0 is selcted and column index 1 is selected
```

```python
np.unique(y,return_counts=True)
```

```python
#Final Visualisation
plt.figure(figsize=(10,5))
for i in range(k):
  plt.scatter(x[y==i,0],x[y==i,1],label=f'Cluster {i}')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],s=300,c='yellow',label='Centroids')
plt.legend()
```
